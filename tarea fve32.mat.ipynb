```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import networkx as nx
import seaborn as sns
```

```python
path = r"fve32.mat"
mat_json = scipy.io.loadmat(path)
def open_mat_file(path: str):
    mat_json = scipy.io.loadmat(path)
    keys = list(mat_json.keys())
    lista = []
    for idx in range(3, len(mat_json.keys())) :
        lista.append(mat_json[keys[idx]])
    return mat_json, lista
json, [names, CIJ] = open_mat_file('fve32.mat')
sns.heatmap(CIJ)
```

```python
G = nx.from_numpy_array(CIJ) 
nx.draw(G, with_labels=True)
```

```python
plt.figure(figsize=(16, 8))
plt.subplot(2,2,1)
nx.draw(G, pos=nx.spring_layout(G), with_labels=True, node_color='yellow')
plt.subplot(2,2,2)
nx.draw(G, pos=nx.circular_layout(G), with_labels=True, node_color='hotpink')
plt.subplot(2,2,3)
nx.draw(G, pos=nx.shell_layout(G), with_labels=True, node_color='magenta')
plt.subplot(2,2,4)
nx.draw(G, pos=nx.kamada_kawai_layout(G), with_labels=True, node_color='cyan') 
```
