```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import networkx as nx
import seaborn as sns
```

```python
path = r"Coactivation_matrix.mat"
mat_json = scipy.io.loadmat(path)
def open_mat_file(path: str):
    mat_json = scipy.io.loadmat(path)
    keys = list(mat_json.keys())
    lista = []
    for idx in range(3, len(mat_json.keys())) :
        lista.append(mat_json[keys[idx]])
    return mat_json, lista
json, [CM, coord] = open_mat_file('Coactivation_matrix.mat')
sns.heatmap(CM)
```

```python
df=CM[9:29,9:29]
df = pd.DataFrame(df)
sns.heatmap(df)
```

```python
G = nx.from_pandas_adjacency(df, create_using=nx.DiGraph())
nx.draw_networkx(G, arrows=True)
```

```python
grados = dict(G.degree())
g = pd.DataFrame(grados.values())
max(grados, key=grados.get)
```

```python
plt.hist(g, bins = 20)
g.describe()
```
