```python
import numpy as np
import matplotlib.pyplot as plt
# semilla para num aleatorio [-3,3]
np.random.seed(5002) 
ej1 = np.random.randint(-3, 4)
```

```python
np.random.seed(5002)
ej2 = np.random.uniform(-10, 10)
```

```python
def analizar_distribucion(mu, sigma, total_numeros): 
    numeros = np.random.normal(mu, sigma, total_numeros) # array
    plt.hist(numeros, bins=50)
    media= np.mean(numeros)
    sigma = np.std(numeros)
    rango_1 = ((media - sigma <= numeros) & (numeros <= media + sigma)).mean() * 100
    rango_2 = ((media - 2 * sigma <= numeros) & (numeros <= media + 2 * sigma)).mean() * 100
    rango_3 = ((media - 3 * sigma <= numeros) & (numeros <= media + 3 * sigma)).mean() * 100
    print('1er cuartil ', rango_1, ' 2er cuartil ', rango_2, ' 3er cuartil ', rango_3)

analizar_distribucion(ej1,ej2,1000)
```

```python
analizar_distribucion(0, 1., 1000)
```

```python
analizar_distribucion(3, 5, 5000)
```

```python
analizar_distribucion(-3, 3, 10000)
```
